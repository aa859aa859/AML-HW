```python
import pandas as pd
import plotly as plt
import matplotlib.pyplot as plt
import seaborn as sns
filepath = "/home/achal/Desktop/Cornell/AML/HW1/iris.data"
```

```python
_col_labels = ['sepal length', 'sepal width','petal length','petal width','class']
#all numeric measurements in cm
df = pd.read_csv(filepath, sep=",",header=None)
df.columns = _col_labels
print(df) #answer to number 2 part 1 creating Nxp array
```
