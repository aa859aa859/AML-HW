```python
%matplotlib inline
import imageio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
#1a) Load dataset
train_labels, train_data = [], []
test_labels, test_data = [], []

with open('./faces/train.txt') as f:
    for line in f:
        im = imageio.imread(line.strip().split()[0])
        train_data.append(im.reshape(2500,))
        train_labels.append(line.strip().split()[1])
        
with open('./faces/test.txt') as f:
    for line in f:
        im = imageio.imread(line.strip().split()[0])
        test_data.append(im.reshape(2500,))
        test_labels.append(line.strip().split()[1])
#1b) Load training set   
train_data, train_labels = np.array(train_data, dtype=float), np.array(train_labels, dtype=int)
test_data, test_labels = np.array(test_data, dtype=float), np.array(test_labels, dtype=int)
print("training data (shape, lables)- ", train_data.shape, train_labels.shape)
print("testing data (shape, lables) - ", test_data.shape, test_labels.shape)
```

```python
#1b) Draw a face
#Training face
plt.subplot(1, 2, 1)
plt.imshow(train_data[5, :].reshape(50,50), cmap = cm.Greys_r)
plt.title('Train sample Image')
#testing face
plt.subplot(1, 2, 2)
plt.imshow(test_data[5, :].reshape(50,50), cmap = cm.Greys_r)
plt.title('Test sample Image')
plt.show()
```

```python
#1c) Draw average face.
avg = np.apply_along_axis(np.mean, 0, train_data)
plt.imshow(avg.reshape(50,50), cmap = cm.Greys_r) #grayscale
plt.title('Average face ($\mu$)')
plt.show()
print('Average face shape = ', avg.shape)
```

```python
#1d) Mean substraction
train_ms = train_data - avg
test_ms = test_data - avg

#Plot faces
plt.figure(figsize=(6, 6))
plt.subplot(2,2,1)
plt.imshow(train_data[5, :].reshape(50,50), cmap = cm.Greys_r)
plt.title("Train sample")

plt.subplot(2,2,2)
plt.imshow(train_ms[5,:].reshape(50,50), cmap = cm.Greys_r)
plt.title("Train sample - $\mu$")

plt.subplot(2,2,3)
plt.imshow(test_data[5,:].reshape(50,50), cmap = cm.Greys_r)
plt.title("Test sample")

plt.subplot(2,2,4)
plt.imshow(test_ms[5,:].reshape(50,50), cmap = cm.Greys_r)
plt.title("Test sample - $\mu$")
plt.show()
```

```python
# 1e) Eigenface - SVD decomposition
u, s, v = np.linalg.svd(train_ms)
print("SVD decomposition u,s,v - ", u.shape, s.shape, v.shape)

#plotting first 10 eigenface 
plt.figure(figsize=(10, 6))
for i in range(10):
    plt.subplot(3, 5, i + 1)
    plt.imshow(v[i,:].reshape(50,50), cmap = cm.Greys_r)
    plt.title('eigenface %d' % (i+1))
    
#plotting last five eigen face
for j in range(5):
    plt.subplot(3, 5, j + 11)
    plt.imshow(v[j - 5, :].reshape(50,50), cmap = cm.Greys_r)
    plt.title('Last eigenface %d' % (5-j))
        
plt.tight_layout()
plt.show()
```

```python
#1f) low-Rank Approximation
def generate_lr(r):
    return u[:, :r].dot(np.diag(s[:r]).dot(v[:r, :]))

#plot rank approximation error
lr_error = []
for r in range(200):
    lr_error.append(np.linalg.norm(generate_lr(r + 1) - train_data, ord='fro'))
    
#plot
plt.figure(figsize=(6,4))
plt.plot(range(1, 201), lr_error, linewidth=4)
plt.xticks(range(0, 201, 25))
plt.title("Low rank error")
plt.xlabel('Rank-r')
plt.ylabel('Error')
plt.grid()
plt.tight_layout()
```

```python
# 1g) Eigenface Feature
# eigen matr: p * p
# Input matrix - training/testing : N * p
def get_ef_feature(ev, trainingM, testingM, rank=1):
    return trainingM.dot(ev[:rank, :].T), testingM.dot(ev[:rank, :].T)
```

```python
# 1h) Face Recognition
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

    
accuracy = []
for r in range(1, 201):
    train, test = get_ef_feature(v, train_ms, test_ms, r)
    #Convergence issue - try with different solvers (not helped), computer lags on increasing max_iter
    model = LogisticRegression(solver='liblinear') 
    model.fit(train, train_labels)
    accuracy.append(model.score(test, test_labels))
    
print(len(accuracy))
```

```python
#plotting accuracy
plt.figure(figsize=(6, 4))
plt.plot(range(1,201), np.array(accuracy) * 100, linewidth=4)
plt.xticks(range(0, 201, 25))
plt.xlabel("Rank-r")
plt.ylabel("Accuracy(%)")
plt.title("Accuracy with - r", fontsize=16)
plt.grid()
plt.show()
```
